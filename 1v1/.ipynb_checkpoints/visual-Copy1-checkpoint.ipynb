```python
import pandas as pd
import glob
import os
from tqdm import tqdm

ts_dir = "ts_data"

files = glob.glob(f"{ts_dir}/ts_seed_*.parquet")
print(f"Znaleziono {len(files)} plików.")

results = []

for f in tqdm(files, desc="Przetwarzanie"):
    seed = int(os.path.basename(f).split('_')[-1].split('.')[0])

    # wczytaj cały plik (500k wierszy)
    df = pd.read_parquet(f)

    avg_price1  = df["p1"].mean()
    avg_price2  = df["p2"].mean()
    avg_price   = 0.5 * (avg_price1 + avg_price2)

    avg_profit1 = df["pi1"].mean()
    avg_profit2 = df["pi2"].mean()
    avg_profit  = 0.5 * (avg_profit1 + avg_profit2)

    market_price = df["market_price"].mean()

    results.append({
        "seed": seed,
        "avg_price1": avg_price1,
        "avg_price2": avg_price2,
        "avg_price": avg_price,
        "avg_profit1": avg_profit1,
        "avg_profit2": avg_profit2,
        "avg_profit": avg_profit,
        "market_price": market_price
    })

# zamiana na DataFrame i zapis
df_res = pd.DataFrame(results)
df_res.to_csv("analysis_results_simple.csv", index=False)

print("Zapisano analysis_results_simple.csv")
```

```python
import numpy as np
import pandas as pd
import glob
import os
from tqdm import tqdm

T = 500000
ts_dir = "ts_data"

files = glob.glob(f"{ts_dir}/ts_seed_*.parquet")
n_seeds = len(files)

print("Znaleziono", n_seeds, "plików.")

# sumy po seedach
pi1_sum = np.zeros(T)
pi2_sum = np.zeros(T)
p1_sum = np.zeros(T)
p2_sum = np.zeros(T)
market_sum = np.zeros(T)

# ======== PĘTLA PO SEEDACH ========
for f in tqdm(files, desc="Przetwarzanie"):

    df = pd.read_parquet(f)

    pi1_sum += df["pi1"].values
    pi2_sum += df["pi2"].values
    p1_sum += df["p1"].values
    p2_sum += df["p2"].values
    market_sum += df["market_price"].values

# ======== UŚREDNIENIE ========
pi1_avg = pi1_sum / n_seeds
pi2_avg = pi2_sum / n_seeds
profit_avg = (pi1_avg + pi2_avg) / 2

p1_avg = p1_sum / n_seeds
p2_avg = p2_sum / n_seeds
market_avg = market_sum / n_seeds

# ======== ZŁOŻENIE DATAFRAME ========
df_out = pd.DataFrame({
    "t": np.arange(T),
    "avg_profit1": pi1_avg,
    "avg_profit2": pi2_avg,
    "avg_profit": profit_avg,
    "avg_price1": p1_avg,
    "avg_price2": p2_avg,
    "avg_market_price": market_avg
})

df_out.to_csv("analysis_results_timeseries.csv", index=False)

print("Zapisano analysis_results_timeseries.csv")
```

```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from tqdm import tqdm
import os

# ========== PARAMETRY ==========
step = 1000
T = 500000
t_points = np.arange(0, T, step) / 1000
ts_dir = "ts_data"
n_clusters = 4
n_samples_per_cluster = 20
output_dir = "cluster_plots_fullset"
os.makedirs(output_dir, exist_ok=True)

# Benchmarks
profit_bench = 0.0611
profit_mono = 0.125
price_bench = 0.15
price_mono = 0.5

# ========= 6 ZMIENNYCH DO KLASTERYZACJI ==========
variables = {
    "avg_profit":  ("Śr. zysk (mean)",     profit_bench, profit_mono),
    "avg_profit1": ("Śr. zysk firm 1",     profit_bench, profit_mono),
    "avg_profit2": ("Śr. zysk firm 2",     profit_bench, profit_mono),
    "avg_price":   ("Śr. cena (mean)",     price_bench,  price_mono),
    "avg_price1":  ("Cena firm 1",         price_bench,  price_mono),
    "avg_price2":  ("Cena firm 2",         price_bench,  price_mono),
}

# ========= WCZYTYWANIE PLIKÓW ==========
files = sorted(glob.glob(f"{ts_dir}/ts_seed_*.parquet"))
n_seeds = len(files)
print("Znaleziono", n_seeds, "plików.")

# ========= FUNKCJA KLASTERYZACJI ==========
def cluster_variable(var_name, title, bench, mono):

    print(f"\n=== KLASTERYZACJA: {var_name} ===")

    # przygotuj macierz trajektorii
    trajectories = np.zeros((n_seeds, len(t_points)))

    # ------ wczytanie trajektorii dla tej zmiennej ------
    for i, f in enumerate(tqdm(files, desc=f"Wczytywanie {var_name}")):
        df = pd.read_parquet(f)
        df_s = df.iloc[::step]

        if "profit" in var_name:
            if var_name == "avg_profit":
                x = (df_s["pi1"].values + df_s["pi2"].values) / 2
            elif var_name == "avg_profit1":
                x = df_s["pi1"].values
            elif var_name == "avg_profit2":
                x = df_s["pi2"].values

        else:  # ceny
            if var_name == "avg_price":
                x = (df_s["p1"].values + df_s["p2"].values) / 2
            elif var_name == "avg_price1":
                x = df_s["p1"].values
            elif var_name == "avg_price2":
                x = df_s["p2"].values

        trajectories[i, :] = x

    trajectories = np.nan_to_num(trajectories)

    # ------ KLASTERYZACJA ------
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(trajectories)

    # ilość w grupach
    uniq, cnt = np.unique(labels, return_counts=True)
    for u, c in zip(uniq, cnt):
        print(f"  grupa {u}: n={c}")

    colors = plt.cm.tab10(np.linspace(0, 1, n_clusters))

    # ------ WYKRESY per CLUSTER ------
    for label in uniq:
        mask = labels == label
        group = trajectories[mask]
        n_group = len(group)

        mean_traj = group.mean(axis=0)
        std_traj = group.std(axis=0)

        plt.figure(figsize=(12,6))

        # losowe sample
        sample_idx = np.random.choice(n_group, size=min(n_samples_per_cluster, n_group), replace=False)
        for idx in sample_idx:
            plt.plot(t_points, group[idx], color=colors[label], alpha=0.12, linewidth=0.6)

        # pasmo ± std
        plt.fill_between(t_points, mean_traj - std_traj, mean_traj + std_traj,
                         color=colors[label], alpha=0.22)

        # średnia
        plt.plot(t_points, mean_traj, color="black", linewidth=2.4, label="Średnia grupy")

        # linie benchmarkowe
        plt.axhline(bench, color="green", linestyle="--", label=f"Benchmark ({bench})")
        plt.axhline(mono, color="orange", linestyle="--", label=f"Monopol ({mono})")

        plt.title(f"{title} – grupa {label} (n={n_group})")
        plt.xlabel("t (tysiące)")
        plt.ylabel(title)
        plt.grid(True, alpha=0.3)
        plt.legend()
        plt.tight_layout()

        fname = f"{output_dir}/{var_name}_cluster_{label}.png"
        plt.savefig(fname, dpi=150)
        plt.close()
        print("  zapisano:", fname)

# ========= URUCHOM WSZYSTKIE 6 KLASTERYZACJE ==========
for var, (title, bench, mono) in variables.items():
    cluster_variable(var, title, bench, mono)

print("\nGOTOWE: wygenerowano klastry dla 6 zmiennych.")
```

```python
import numpy as np
import pandas as pd

def cluster_summary(trajectories, labels, t_points):
    """
    trajectories: shape (n_seeds, n_times)
    labels: array (n_seeds)
    t_points: vector of times (important for indexing)
    """

    summary_rows = []

    for cl in np.unique(labels):
        mask = labels == cl
        group = trajectories[mask]

        # średnia trajektoria
        mean_traj = group.mean(axis=0)

        # early (10k), mid (100k), late (500k)
        early = mean_traj[10] if len(mean_traj) > 10 else mean_traj[0]
        mid   = mean_traj[len(mean_traj)//2]
        late  = mean_traj[-1]

        # trend (slope)
        slope = late - early

        # stabilność
        std_global = group.std()

        # "cyklowatość"
        amplitude = (group.max(axis=1) - group.min(axis=1)).mean()

        summary_rows.append({
            "cluster": cl,
            "n": group.shape[0],
            "value_early": early,
            "value_mid": mid,
            "value_late": late,
            "trend_slope": slope,
            "std": std_global,
            "amplitude": amplitude
        })

    return pd.DataFrame(summary_rows)
```

```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from tqdm import tqdm
import os

# ========== PARAMETRY ==========
step = 1000
T = 500000
t_points = np.arange(0, T, step) / 1000
ts_dir = "ts_data"
n_clusters = 4
n_samples_per_cluster = 20
output_dir = "cluster_plots_fullset"
os.makedirs(output_dir, exist_ok=True)

# Benchmarks
profit_bench = 0.0611
profit_mono = 0.125
price_bench = 0.15
price_mono = 0.5

# ========= 6 ZMIENNYCH DO KLASTERYZACJI ==========
variables = {
    "avg_profit":  ("Śr. zysk (mean)",     profit_bench, profit_mono),
    "avg_profit1": ("Śr. zysk firm 1",     profit_bench, profit_mono),
    "avg_profit2": ("Śr. zysk firm 2",     profit_bench, profit_mono),
    "avg_price":   ("Śr. cena (mean)",     price_bench,  price_mono),
    "avg_price1":  ("Cena firm 1",         price_bench,  price_mono),
    "avg_price2":  ("Cena firm 2",         price_bench,  price_mono),
}

# ========= WCZYTYWANIE PLIKÓW ==========
files = sorted(glob.glob(f"{ts_dir}/ts_seed_*.parquet"))
n_seeds = len(files)
print("Znaleziono", n_seeds, "plików.")


# ========= FUNKCJA TABELI PODSUMOWUJĄCEJ ==========
def cluster_summary(trajectories, labels):
    rows = []
    for cl in np.unique(labels):
        group = trajectories[labels == cl]
        mean_traj = group.mean(axis=0)
        
        early = mean_traj[5]                    # ~5k
        mid   = mean_traj[len(mean_traj)//2]    # ~250k
        late  = mean_traj[-1]                   # ~500k

        slope = late - early
        std_global = group.std()
        amplitude = (group.max(axis=1) - group.min(axis=1)).mean()

        rows.append({
            "cluster": cl,
            "count": group.shape[0],
            "early_value": round(early, 5),
            "mid_value": round(mid, 5),
            "late_value": round(late, 5),
            "trend": round(slope, 5),
            "std": round(std_global, 5),
            "amplitude": round(amplitude, 5),
        })

    return pd.DataFrame(rows)


# ========= GŁÓWNA FUNKCJA KLASTERYZACJI ==========
def cluster_variable(var_name, title, bench, mono):

    print(f"\n=== KLASTERYZACJA: {var_name} ===")

    # przygotuj macierz trajektorii
    trajectories = np.zeros((n_seeds, len(t_points)))

    for i, f in enumerate(tqdm(files, desc=f"Wczytywanie {var_name}")):
        df = pd.read_parquet(f)
        df_s = df.iloc[::step]

        if "profit" in var_name:
            if var_name == "avg_profit":
                x = (df_s["pi1"].values + df_s["pi2"].values) / 2
            elif var_name == "avg_profit1":
                x = df_s["pi1"].values
            elif var_name == "avg_profit2":
                x = df_s["pi2"].values

        else:
            if var_name == "avg_price":
                x = (df_s["p1"].values + df_s["p2"].values) / 2
            elif var_name == "avg_price1":
                x = df_s["p1"].values
            elif var_name == "avg_price2":
                x = df_s["p2"].values

        trajectories[i, :] = x

    trajectories = np.nan_to_num(trajectories)

    # ==== KLASTERYZACJA ====
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(trajectories)

    print("Rozkład klastrów:", dict(zip(*np.unique(labels, return_counts=True))))

    colors = plt.cm.tab10(np.linspace(0, 1, n_clusters))

    # ==== RYSOWANIE I TABELKI ====
    summary_df = cluster_summary(trajectories, labels)
    
    print("\nTABELA PODSUMOWUJĄCA DLA:", var_name)
    print(summary_df, "\n")

    # zapis tabeli
    summary_df.to_csv(f"{output_dir}/{var_name}_summary.csv", index=False)

    # osobne wykresy
    for cl in np.unique(labels):
        group = trajectories[labels == cl]
        n_group = len(group)
        mean_traj = group.mean(axis=0)
        std_traj = group.std(axis=0)

        plt.figure(figsize=(12, 6))

        # losowe sample
        idx = np.random.choice(n_group, size=min(n_samples_per_cluster, n_group), replace=False)
        for i in idx:
            plt.plot(t_points, group[i], color=colors[cl], alpha=0.12, linewidth=0.6)

        # pasmo ± std
        plt.fill_between(t_points, mean_traj - std_traj,
                         mean_traj + std_traj,
                         color=colors[cl], alpha=0.22)

        plt.plot(t_points, mean_traj, color="black", linewidth=2.4, label="Średnia")

        plt.axhline(bench, color="green", linestyle="--", label=f"Benchmark ({bench})")
        plt.axhline(mono, color="orange", linestyle="--", label=f"Monopol ({mono})")

        plt.title(f"{title} – klaster {cl} (n={n_group})")
        plt.xlabel("t (tys.)")
        plt.ylabel(title)
        plt.grid(True, alpha=0.3)
        plt.legend()
        plt.tight_layout()

        fname = f"{output_dir}/{var_name}_cluster_{cl}.png"
        plt.savefig(fname, dpi=150)
        
        plt.show()   # <<==== WYŚWIETLANIE W JUPYTERZE
        print("Zapisano:", fname)
        
        # WYŚWIETLENIE tabelki POD wykresem
        display(summary_df[summary_df["cluster"] == cl])


# ========= URUCHOM WSZYSTKO ==========
for var, (title, bench, mono) in variables.items():
    cluster_variable(var, title, bench, mono)

print("\nGOTOWE.")
```

```python
import numpy as np
import pandas as pd
import os
from tqdm import tqdm

T = 500_000
step = 25_000
checkpoints = list(range(0, T, step)) + [T-1]

k = 6
n = k + 1

ts_dir = "ts_data"
q_dir = "q_tables"

files = sorted([f for f in os.listdir(ts_dir) if f.startswith("ts_seed_")])

rows = []

for fname in tqdm(files, desc="Seeds"):
    seed = int(fname.split("_")[-1].split(".")[0])

    # --- load trajectory and Q tables ---
    df = pd.read_parquet(f"{ts_dir}/{fname}")
    Q1 = np.load(f"{q_dir}/Q1_seed_{seed}.npy")
    Q2 = np.load(f"{q_dir}/Q2_seed_{seed}.npy")   # ndim=2 or 3

    # --- loop over checkpoints ---
    for t in checkpoints:

        # --- extract last available state ---
        if t < 2:   # need at least 3 points for memory2 agent
            continue

        p1 = df["p1"].values[:t+1]
        p2 = df["p2"].values[:t+1]

        # actions at time t
        a1 = int(round(p1[-1] * k))
        a2 = int(round(p2[-1] * k))

        # state for agent 1 (memory=1)
        s1 = int(round(p2[-2] * k))

        # state for agent 2 (memory=2)
        s2_1 = int(round(p1[-2] * k))
        s2_2 = int(round(p1[-3] * k))

        # --- optimality gamma1 ---
        q1_sa = Q1[s1, a1]
        q1_best = Q1[s1].max()
        gamma1 = q1_sa / q1_best if q1_best > 0 else 0

        # --- optimality gamma2 ---
        if Q2.ndim == 3:
            q2_sa = Q2[s2_1, s2_2, a2]
            q2_best = Q2[s2_1, s2_2].max()
        else:
            q2_sa = Q2[a2]
            q2_best = Q2.max()

        gamma2 = q2_sa / q2_best if q2_best > 0 else 0

        # --- Nash criterion ---
        is_nash = (gamma1 > 0.9999) and (gamma2 > 0.9999)

        rows.append({
            "seed": seed,
            "t": t,
            "gamma1": gamma1,
            "gamma2": gamma2,
            "is_nash": is_nash
        })

df = pd.DataFrame(rows)
df.to_csv("optimality_over_time.csv", index=False)

print("\nSaved: optimality_over_time.csv")
```
