```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytaj dane
df = pd.read_csv("figure1_data.csv")  # upewnij się, że plik jest w bieżącym folderze

# Wykres 1: Średnia optymalność Γ
plt.figure(figsize=(10, 5))
plt.plot(df["T"], df["avg_gamma"], marker='o', linestyle='-', color='b', label='Γ (Twoje dane)')
plt.xlabel("Całkowity czas uczenia T")
plt.ylabel("Średnia optymalność Γ")
plt.title("Średnia optymalność w funkcji T")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig("gamma_plot.png", dpi=150)
plt.show()

# Wykres 2: Udział równowag Nasha
plt.figure(figsize=(10, 5))
plt.plot(df["T"], df["share_nash"], marker='s', linestyle='-', color='r', label='Udział Nasha (Twoje dane)')
plt.xlabel("Całkowity czas uczenia T")
plt.ylabel("Udział równowag Nasha")
plt.title("Udział równowag Nasha w funkcji T")
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.savefig("nash_plot.png", dpi=150)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Wczytaj dane
df = pd.read_csv("figure1_data.csv")

df0 = pd.DataFrame({"T": [0], "avg_gamma": [0.0], "share_nash": [0.0]})
df = pd.concat([df0, df], ignore_index=True).sort_values("T")

# --- Wykres 1: Średnia optymalność Γ ---
plt.figure(figsize=(10, 5))
plt.plot(df["T"], df["avg_gamma"], marker='o', linestyle='-', color='b', label='Γ')
plt.xlabel("Całkowity czas uczenia T")
plt.ylabel("Średnia optymalność Γ")
plt.title("Średnia optymalność w funkcji T")
plt.grid(True, alpha=0.3)
plt.legend(loc='lower right')
plt.xlim(left=0)                  
plt.xticks(df["T"], rotation=45)     
plt.ylim(bottom=0, top=1.05)
plt.tight_layout()
plt.savefig("gamma_plot.png", dpi=150)
plt.show()

# --- Wykres 2: Udział równowag Nasha (w %) ---
plt.figure(figsize=(10, 5))
# Konwersja ułamków na procenty
nash_percent = df["share_nash"] * 100
plt.plot(df["T"], nash_percent, marker='s', linestyle='-', color='r', label='Udział Nasha')
plt.xlabel("Całkowity czas uczenia T")
plt.ylabel("Udział równowag Nasha (%)")
plt.title("Udział równowag Nasha w funkcji T")
plt.grid(True, alpha=0.3)
plt.legend(loc='lower right')
plt.xlim(left=0)                     # oś x od 0
plt.xticks(df["T"], rotation=45)     # wszystkie wartości T na osi
plt.ylim(bottom=0, top=100)
plt.tight_layout()
plt.savefig("nash_plot.png", dpi=150)
plt.show()
```
